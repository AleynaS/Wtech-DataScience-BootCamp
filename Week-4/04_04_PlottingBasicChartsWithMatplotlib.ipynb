1. Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)

df = df[df['continent'] == df['continent']]
totalCases = df.loc[:,['location','total_cases']]



totalCases = totalCases.groupby(by = "location").max()
totalCases = totalCases.sort_values(by='total_cases', ascending=False)
totalCases = totalCases.iloc[0:20]
totalCases

fig = plt.figure(figsize=(28, 5))
plt.bar(totalCases.index,totalCases['total_cases'])
plt.show()
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
dailyDeaths = df[df['location']=='Turkey']['new_deaths']

fig = plt.figure()

plt.subplot(1, 3, 1)
plt.hist(dailyDeaths,bins=10)
plt.show()

plt.subplot(1, 3, 2)
plt.hist(dailyDeaths,bins=15)
plt.show()

plt.subplot(1, 3, 3)
plt.hist(dailyDeaths,bins=30)
plt.show()
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
franceDf = df[df['location']=='France' ]
germanyDf =df[df['location']=='Germany']

plt.figure(figsize=(25, 10))
plt.subplot(2, 2, 1)
plt.title("france new cases")
plt.scatter(franceDf['date'],franceDf['new_cases'])
plt.show()

plt.figure(figsize=(25, 10))
plt.subplot(2, 2, 2)
plt.title("france new deaths")
plt.scatter(franceDf['date'],franceDf['new_deaths'])
plt.show()

plt.figure(figsize=(25, 10))
plt.subplot(2, 2, 3)
plt.title("germany new cases")
plt.scatter(germanyDf['date'],germanyDf['new_cases'])
plt.show()

plt.figure(figsize=(25, 10))
plt.subplot(2, 2, 4)
plt.title("germany new deaths")
plt.scatter(germanyDf['date'],germanyDf['new_deaths'])
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
dailyDeaths = df[df['location']=='Turkey']['new_deaths']

plt.figure(figsize=(20, 6))
plt.title('Turkey daily deaths box plot')
plt.boxplot(dailyDeaths)
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart 

```python
df = df.groupby(by = "continent").sum()



plt.pie(df['new_cases'], labels=df.index, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
