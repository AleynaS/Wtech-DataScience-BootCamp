1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)


df = df[df['location'].isin(['Spain', 'France', 'Germany', 'Italy']) & (df['date']>"04/01/2020")]
spainDf = df[df['location']=='Spain']
franceDf = df[df['location']=='France']
germanyDf = df[df['location']=='Germany']
italyDf = df[df['location']=='Italy']

plt.figure(figsize=(20, 5))
sns.lineplot(x = spainDf['date'], y = spainDf['total_deaths'], data = spainDf, lw = 3, color = 'r')
sns.lineplot(x = franceDf['date'], y = franceDf['total_deaths'], data = franceDf, lw = 3, color = 'g')
sns.lineplot(x = germanyDf['date'], y = germanyDf['total_deaths'], data = germanyDf, lw = 3, color = 'b')
sns.lineplot(x = italyDf['date'], y = italyDf['total_deaths'], data = italyDf, lw = 3, color = 'y')
plt.legend(['Spain', 'France', 'Germany', 'Italy'])
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df = df[df['continent'] == df['continent']]
df = df.groupby(by = "continent").sum()
print(type(df))

plt.figure(figsize=(12,4))
sns.barplot(x= df.index, y = df['new_deaths'], data = df)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
dailyDeaths = df[df['location']=='Turkey']
dailyDeaths

plt.figure(figsize=(20, 15))

plt.subplot(2,2,1)
sns.distplot(dailyDeaths['new_deaths'], bins=50, kde = False, color = '#22B573')
plt.title("New deaths\nBins=50/kde = False")

plt.subplot(2,2,2)
sns.distplot(dailyDeaths['new_deaths'], bins=20, kde = False, color = '#22B573')
plt.title("New deaths\nBins=20/kde = False")

plt.subplot(2,2,3)
sns.distplot(dailyDeaths['new_deaths'], bins=50, color = 'b')
plt.title("New deaths\nBins=50/kde = True")

plt.subplot(2,2,4)
sns.distplot(dailyDeaths['new_deaths'], bins=20, color = '#FF914D')
plt.title("New deaths\nBins=20/kde = True")

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)


df = df[df['location'].isin(['Spain', 'France'])]
spainDf = df[df['location']=='Spain']
franceDf = df[df['location']=='France']

plt.figure(figsize=(25, 6))

plt.subplot(1,3,1)
sns.boxplot(x = 'location',  y = 'new_deaths', data = df, palette = ['#FF914D', '#22B573'])

plt.subplot(1,3,2)
sns.violinplot(x = 'location',  y = 'new_deaths', data = df, palette = ['#22B573', '#FF914D'])

plt.subplot(1,3,3)
sns.swarmplot(x = 'location', y = 'new_deaths', data=df)

plt.show()
```
